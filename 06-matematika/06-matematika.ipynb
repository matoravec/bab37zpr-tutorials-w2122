## 05.01: Gaussova eliminačná metóda

```python
m = [
    [2, 1, -1, 8],
    [-3, -1, 2, -11],
    [-2, 1, 2, -3]
]
```

```python
pivot_r = 0
pivot_c = 0
r_total = len(m)
c_total = len(m[0])

while pivot_r < r_total and pivot_c < c_total:
    if all([m[pivot_c][r] == 0 for r in range(r_total)]):
        pivot_c += 1
    else:
        column = [abs(m[r][pivot_c]) for r in range(r_total)]
        argmax_c = column.index(max(column))
        if argmax_c != pivot_r:
            m[pivot_r], m[argmax_c] = m[argmax_c], m[pivot_r] # Row swap
        for r in range(pivot_r + 1, r_total):
            factor = m[r][pivot_c] / m[pivot_r][pivot_c]
            print(m[r][pivot_c], m[pivot_r][pivot_c], factor)
            m[r][pivot_c] = 0
            for c in range(pivot_c + 1, c_total):
                print(f'r:{r}, c:{c}')
                m[r][c] = m[r][c] - m[pivot_r][c] * factor
                
            pivot_c += 1
            pivot_r += 1
        print(m)
```
